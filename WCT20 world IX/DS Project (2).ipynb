```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
import json
```

**MATCH SUMMARY**

```python
#Open JSON file and make a dataframe out of it
with open('t20_wc_match_results.json') as f:
    data=json.load(f)

df_match=pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
#Change 'scorecard' to 'matchid'
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match.head()
```

```python
#Dictonary to map match id in Batting DataFrame
mid_dict={}
for index,row in df_match.iterrows():
    k1=row['team1']+' Vs '+row['team2']
    k2=row['team2']+' Vs '+row['team1']
    
    mid_dict[k1]=row['match_id']
    mid_dict[k2]=row['match_id']
    
mid_dict
```

**Batting Summary**

```python
#Open JSON file and make a dataframe out of it
with open('t20_wc_batting_summary.json') as f:
    data=json.load(f) #Has 45 json elements
    arr=[]
    for rec in data:
        arr.extend(rec['battingSummary'])
df_bat=pd.DataFrame(arr)
```

```python
#Convert Dismissal Column in Out/NotOut
#If dismissal column blank its NotOut
df_bat['Out/NotOut']=df_bat.dismissal.apply(lambda x: 'out' if len(x)>0 else 'NotOut')
df_bat=df_bat.drop('dismissal',axis='columns') #drop Dismissal
df_bat.head(15)
```

```python
#Remove Special Characters from Batsmen Names
df_bat['batsmanName']=df_bat['batsmanName'].apply(lambda x: x.replace('â€',''))
df_bat['batsmanName']=df_bat['batsmanName'].apply(lambda x: x.replace('\xa0',''))

df_bat.head(15)
```

```python
#Map MAtch ID
df_bat['match_id']= df_bat['match'].map(mid_dict)
df_bat.head(8)
```

**BOWLING SUMMARY**

```python
with open('t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowl = pd.DataFrame(all_records)
df_bowl.head()
```

```python
#Map match ID
df_bowl['match_id'] = df_bowl['match'].map(mid_dict)
df_bowl.head()
```

**Player Summary**

```python
with open('t20_wc_player_info.json') as f:
    data = json.load(f)
```

```python
df_players = pd.DataFrame(data)

print(df_players.shape)
df_players.head(10)
```

```python
#Remove Special Chracters
df_players['name'] = df_players['name'].apply(lambda x: x.replace('â€', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('†', ''))
df_players['name'] = df_players['name'].apply(lambda x: x.replace('\xa0', ''))
df_players.head(10)
```
